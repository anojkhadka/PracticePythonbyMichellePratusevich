Generate a random number between 1 and 9 (including 1 and 9). Ask the user to guess the number, then tell them whether they guessed too low, too high, or exactly right.

<br><br>

Extras:

<ul>
<li> Keep the game going until the user types “exit”
<li> Keep track of how many guesses the user has taken, and when the game ends, print this out.
</ul>
</font>

<font size = 2, color = green>Brief Description of Code</font>
<br><br>

This code asks user to guess a number and will end and display the result of the game if user either chooses to 'exit' or makes a correct guess. Else, game will keep going. Random module from python has been called to generate a random integer. This random integer generated will not be visible to the player as there is no point in knowing the guess you are going to make. Correct guess will be decided based on the difference between the random integer generated and guess made by the user. If difference is 0, guess is correct. Far guess is displayed if this difference is greater than three and if less than three guess will be assumed to be very near one. 

I have tried to do the error handling in this code without using the 'try .....except'  block of python. If user enters anything other than integers between 1 to 9 included and 'exit' this code will return message and ask to enter input again. Python fucntion is introduced to print the play statistics message every time the game ends. 

```python
import random

random_number = random.randint(1,9)

# These counters are introduced to keep track of the number of times user(or player) performs different 
# activities mentioned.


counter = 0
right_guess_counter = 0
far_guess_counter = 0
near_guess_counter = 0
wrong_entry = 0


# This function will be called to display message and game statistics later on.

def play_statistics():
    print('\n\n')
    print(f'Actual number you were supposed to make a guess: {random_number}\n')
    print(f'Number of times you played the game: {counter}\n')
    print(f'Number of times your guess was too far from the actual number : {far_guess_counter}\n')
    print(f'Number of times your guess was very near to the actual number : {near_guess_counter}\n')
    print(f'Number of times you entered invalid entry: {wrong_entry}')

    
while True:    
        
        user_input = input('Please enter any integer number betweeen 1 and 9 included.'    
                            ' Type "exit", if you want to leave. \n\n')
        
        counter += 1

        
        # Late versions of python treats all input as string. So, I converted integers 1-9 to string to match
        # it with the user input. However, this is not a smart way of doing things when you have a very large
        # number of options.
        
        
        if user_input in [str(1), str(2), str(3), str(4), str(5), str(6), str(7), str(8), str(9)]:


            # To match different conditions below I convert user_input to integer so that it can be treated with
            # mathematical operators.
            
            if (random_number - int(user_input)) == 0:
            
                right_guess_counter += 1
                print('Hurray! You guessed it right.\n')
                print('Check your play statistics below.\n')
                play_statistics()
                break
      
            elif abs(random_number - int(user_input)) > 3:
            
                far_guess_counter += 1
                print('Oops! Guess - too far.\n')
       
            elif abs(random_number - int(user_input)) <= 3:
            
                near_guess_counter += 1
                print('Aha! Guess - very close.\n')
        
        elif user_input == 'exit':
            print('You chose to exit from the game. See you again. Check your play statistics below')
            print('\n')
            play_statistics()
            break
            
        else:
            wrong_entry += 1
            print('Invalid Entry! \n\n')
            

        
        
```
