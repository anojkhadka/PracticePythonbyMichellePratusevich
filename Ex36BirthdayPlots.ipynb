In this code I have used matplotlib python library rather than using bookeh library as said in question. Both are libraries used to visaulize data and preference of one over another is indivisual. To extract data, I have used a simple text file that I created in Ex35. It is a simple text file and will contain all the name of months repeated as many times as there are the number of scientists. This file will be used to create a list of months and which will then be fed to matplotlib command to create a histogram. 

This histogram data will not show the months for which no scientists were born or have 0 value.

```python
from matplotlib import  pyplot as plt
```

```python
with open('Birthday_plot.txt', 'r') as plt_file:
    name = plt_file.read()
    bday_list = name.split()           # Creates a list of all the months name and its value as well.
    

row_data = []                          # Empty list to store only the name of months

for m in range(0, len(bday_list)):     # This loop will go through the 'bday_list' and pick only the name of 
    row_data.append(bday_list[m])      # months to be added in the empty list created before.
    
```

```python
# Below is a functions from matplotlib python library to make a histogram as seen below.


plt.figure(figsize=(8,6), dpi=100)

plt.title('Birthday months statistics of scientists')
plt.xlabel('Name of months')
plt.ylabel('Number of Scientists')
plt.hist(row_data)

plt.show()
```
