This code has been broken into three parts. First part is the function that will ask user to input row and column index. Second part is another function that will take input from the first part and check and fill the 
respective positions accordingly. Last or the third part will run the game and make two players play the game. 

This program does following things.

1. If wrong row column index is entered or player enters the index position that has already been filled, it        will ask the user to enter the right index position again displayed in red.
2. Two players will enter their row column index turn by turn.
3. After each player enters the mark, game status will be shown so that it becomes easier for players to see the 
   available empty positions and the status of the game.
4. Once all the positions have been filled the game automatically stops.



```python
from colorama  import Style, Fore, Back
```

```python
'''This function is created to ask user to input their row and column index for the position where player 
    want to input their tic tac toe mark. This function will later be used and does not take any input. 
    However, it returns the row and column index entered by the player'''


def row_col_input():
        
    row_col = input('Please enter the position \n')
    place_holder = row_col.strip()                      # Changes string to lists and pick row and column index
    m = int(place_holder[0])                            # Gives index position of the row
    n = int(place_holder[2])                            # Gives index position of the column
    return [m,n]
```

```python
''' This function takes the row and column index and will place the tic tac toe mark in that particular position
    This fucntion calls itself(recursive) if player tries to input in the position that has already been filled.
    Once the position is filled, it prints and shows all the filled and empty postion as an output.
    Note: using [x-1][y-1] is because general user does not know python does indexing starting from 0. So,to 
    correct it we did that. '''


def fill_list(x,y):
    
    if sample_list[x-1][y-1] == ' ':       # Checks if there is empty position in the list.
        sample_list[x-1][y-1] = dice       # If position is empty will fill that position with value of 'dice'
    
    elif sample_list[x-1][y-1] != ' ': # If already filled position is entered again this part will corrrect it
        print(Back.RED,'Position already filled! Please choose another position\n')
        print(Style.RESET_ALL)             # From colorama module to print red colored message to warn player
        row_col_index = row_col_input()    # Calling row_col_input() function again to ask for row/column index
        
        fill_list(row_col_index[0],row_col_index[1])     # Recursive call of function to fill the position.    
    
    print('\nGame Status ')                              # Prints the game status in different color so that it
    print(Back.CYAN,f'{sample_list}\n')                  # makes easier for player to see available positions
    print(Style.RESET_ALL)
```
