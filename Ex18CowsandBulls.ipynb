Create a program that will play the “cows and bulls” game with the user. 
    
Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout the game and tell the user at the end.
</font>

<font size = 2 color = green>Brief Description of code</font>
<br><br>
This game is a popular 'cows and bulls' game. Some use 'cows' and 'bulls' in reverse way. However, that does not change the essence of game. Idea about the game has been briefly explained above and one may google(like I did) if have never heard about this game and want to know more.

In context to the code below, it will ask user to enter four unique digits as guess. Digits are not repeated in the number, so one needs to make guess of four different digits number. As stated in the question, each time information about the cows and bulls will be displayed and any wrong entry will be warned with a message. Once the player/user enters a correct guess the game will stop. I have also used '__main__' method and made use of functions. Number to be guessed will be generated with random module that will generate a four unique digits number. 

```python
import random


def cowsandbulls():
    
    '''
    ::: Docstring :::
    
    This is the main function that will do all the work. It does not take any input as an argument and is being 
    called by the main() function. 
    '''
    
    number = random.sample(range(10),4)   # Generates random 4-digit number player needs to guess.
    n = 0                                 # Initiating counter to count the number of times the guess is made.

    while True:
        
        n+=1
        print(f'Play Round {n}')
        
        guess = input(f'Please enter your four digit guess number \n\n')
        
        if guess.isdigit() and len(guess) == 4:  # This loop will checks if wrong entry is made. User is allowed 
                                                 # to enter only 4 digits. Any entry other than 4 digits will       
                                                 # display warning message.
                    
            
            cows = 0        # Initiating counters to count number of 'cows' and 'bulls'                          
            bulls = 0
            
            
            for index in range(0,4):     # Loops through each of the 4 digit number guessed by player to    
                                         # check for'cows' and 'bulls'
                    
                if int(guess[index]) == number[index]:  # Matching elements to check for 'cows'.
                    cows += 1
                    
                elif int(guess[index]) in number:       # Looking for existence of element to check for 'bulls'.
                    bulls += 1

            print(f'\n[cows, bulls] = [{cows}, {bulls}]\n')  # After each guess number of cows and bulls guessed
                                                             # will be displayed
            
            
            if cows == 4:     # This block will end the game or code if right guess has been made.
                
                print(f'Congratulations! You guessed it correct on your {n}th attempt.')
                break
                
        else:
            print('\n Oops wrong entry! Please enter 4 digit number as your guess. \n')
            


            
def main():
    cowsandbulls()
    
    
if __name__ == '__main__':
    main()
```
