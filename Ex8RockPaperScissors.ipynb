<font size = 2, color = red>Solution First</font>

This code is straight forward and uses conditional statements to check all the possible conditions and display the result accordingly. I feel this code is simple and easy to understand.

```python

print('Please enter any one of the following option: \n\n  \t paper \n \t rock \n \t scissor \n\n')

player1 = (input('First player please choose your choice of option \t')).lower()
print('\n')
player2 = (input('Second player please choose your choice of option \t')).lower()         

option_list = ['rock', 'paper', 'scissor']

if player1 in option_list and player2 in option_list:
    
    if player1  == player2:
        print('\n')
        print('The game is tie')
    
    elif player1 == 'rock':
    
        if player2 == 'scissor':
            print('\n')
            print('Player one wins')
        else:
            print('\n')
            print('Player two wins')

    elif player1 == 'paper':
    
        if player2 == 'scissor':
            print('\n')
            print('Player two wins')
        else:
            print('\n')
            print('Player one wins')
        
    elif player1 == 'scissor':
    
        if player2 == 'paper':
            print('\n')
            print('Player one wins')
        else:
            print('\n')
            print('Player two wins')
        
else:
    print('\n')
    print('Invalid entry! Try Again.')
```

<font size = 2, color = red>Solution Second</font>

This solution will employ 'numpy' module and a technique to store all the possible results of the game in each memory unit (or row column address) of an array. Since there are two players with each able to choose three of the given options, there will be total of nine permutations(or say possible outcomes). Based on which option each player chooses result will be accessed from the array which will store each of the possible outcomes. See the image attached for more clarity.  Array can be created without using numpy module as well but numpy is faster. Advantage of this technique is that this does not require any of the conditional statements. It is very light weight and the code itself is very short.






```python
import numpy as np

print("Please type one of these options: \n\n  \t Rock \n \t Paper \n \t Scissor \n")

option_list = ['rock', 'paper','scissor']

player1 = (input('First player please choose your choice of option \t')).lower()
print('\n')
player2 = (input('Second player please choose your choice of option \t')).lower()        


# 'if...else' statement is used here to handle errors. If user inputs any entry other than directed by the code
# it will display a warning message and ask to play again.

if player1 in option_list and player2 in option_list:
    
    # This dictionary will be used to map the entry input by the user to number 0, 1 and 2 so that indexing of 
    # the array can be done later.
    
    dictionary = {'rock':0, 'paper':1, 'scissor':2}


    '''A (3 by 3) array has been created to store all the nine possible outcomes in it. These results are then
    accessed based on the entry player inputs which will be matched to the location of the possible outcomes.'''


    A = np.array([['Tie', 'Second player wins', 'First player wins'],
                 ['First player wins', 'Tie', 'Second player wins'],
                 ['Second player wins', 'First player wins', 'Tie']])

    print('\n\n')
    print(A[dictionary[player1]] [dictionary[player2]]) # Indexing of the array is done here 
            
else:
    print('\n')
    print('Wrong entry! Please try Again.')
```

<font size = 2, color = red>Solution Third</font>

This solution is in principle same as the second one. The only difference is that this handles error using 'try.....except' block of python code. Here any invalid entry will be warned and message will be displayed to enter valid entry. If user enters invalid entry for three rounds program will stop and asks to restart the game. One can observe and compare second solution with this third solution and find how different approach can be applied to handle input and avoid errors. Though I took lot of time writing this solution I feel second solution is more elegant and simple while doing all the required job at the same time.

```python
import numpy as np

print("Please choose: \n\n  \t 0 for Rock \n \t 1 for Paper \n \t 2 for Scissors \n")


# Loop counter introduced to run this loop only for three times

loop_counter = 3
while loop_counter >=1:      
        
        try:
            player1 = int(input('First player please choose number for your choice of option \t'))
              
            print('\n')

            player2 = int(input('Second player please choose number for your choice of option \t'))
            
            
                        
            # This conditional statement is used to raise an error if user inputs integer other than 0,1 or 2
            
            
            if player1 not in [0,1,2] or player2 not in [0,1,2]:
                raise Exception('Oops Invalid Entry! Please enter any one of the integer 0, 1 or 2.\n') 

            A = np.array([['Tie', 'Second player wins', 'First player wins'],
                          ['First player wins', 'Tie', 'Second player wins'],
                          ['Second player wins', 'First player wins', 'Tie']])

            print('\n\n')
            print(A[int(player1)][int(player2)])  
            break
         
        # This will handle ValueError. If user inputs any entry other than integer error will be raised
        # and asked to enter valid entry i.e. 0, 1 or 2.
        
        except ValueError:
            print('\n')
            print('Oops Invalid Entry! Please enter any one of the integer 0, 1 or 2.\n')
            loop_counter -= 1
            if loop_counter == 0:
                print('\n')
                print('Sorry! invalid entry repeatedly entered  at least three times. Please restart the game.')
        
        # This part will raise error. If user inputs any other integer except 0, 1 or 2,  error will
        # be raised and message will be displayed.

        
        except Exception as error:
            print('\n')
            print(error)
            loop_counter -=1
            if loop_counter == 0:
                print('\n')
                print('Sorry! invalid entry repeatedly entered  at least three times. Please restart the game.')
           
        
        '''Note: 'if loop_counter -= 1'  conditional statement used in both error handle is used to count the 
        number of times the wrong entry has been input and to end the program if user inputs wrong entry
        three times. 
        
        '''
```
