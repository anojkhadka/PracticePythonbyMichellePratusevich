Here I have used nested function to build the game and besides random module I have also used colorma module to get colored message output for the user. Though it makes code longer by few lines, output it gives with colored message helps to make game user friendly. As asked in the question, if user repeats the correct guess it just displays warning message and re-asks user to enter input again. If more than six wrong guess is made game will display message and ask if player want to paly the game again. Picture representation of hangman is done by using as many vertical bars as the number of wrong guesses made.   


```python
import random
from colorama import Back, Fore,Style
```

```python
def Hangman():
    
    def word_pick():
        
        ''' This function will return a randomly picked word from the file and return it'''
        
        with open('sowpods.txt', 'r+') as read_file:        # Opening a file 
        
            number = random.randint(0, 267750)              # Using random module to choose a random number 
                                                        
            word = read_file.read().split('\n')             # Will create a list of all words in the file

            random_word = word[number]                      # Returns word from the list at index position 
  
                                                            # 'number' 
        return random_word


    def list_maker(rand_word_fun):  
    
        '''This function is used later to create three different lists and takes the randomly picked word as an 
           input. Output this function returns are: 'empty_list' will keep track of game showing correctly guessed 
           letter and remaining position to be filled. 'guessed_list' is a set(not list) and will be used to keep 
           track of all rightly guessed letters. Similarly, 'counter_list' will be used to keep track of all 
           rightly and wrongly guessed letters.'''
    
        emptylist = []
        guessed_list = set({})
        counter_list = []
    
        for m in range(0, len(rand_word_fun)):
            emptylist.append('  ')
        return emptylist, guessed_list, counter_list


    def Game_restart():
        
        '''This function is used to make a call to restart game.'''
        
        query = input('If you wanna play again type y else type n. ')
        if query == 'y' or query == 'Y':
            guessed_letter_list.clear()         # Empty the lists so that no record of previous game remains 
            game_status_list.clear()            # when we play next game.
            return Hangman()
        else:
            print('You chose to end the game')


    counter = 0                                 # Counter started to keep track of incorrect guesses
    pick = word_pick()                          # picking a random word from word_pick function
    game_status_list = list_maker(pick)[0]      # Creates list with empty items equal in lenght to the random word
    guessed_letter_list = list_maker(pick)[1]   # Empty set used letter to store rightly guessed letter only once.
    counter_letter_list = list_maker(pick)[2]   # Empty list that will store all correctly guessed letter
    wrong_guess_list = []                       # Empty list that will be used to keep track of wrongly guessed
                                                # letter.

    while counter<6:    # This will prevent player from playing even after six incorrect guesses are made.
        
        print(Back.GREEN, f'{len(game_status_list) - len(counter_letter_list)} more letters to complete the game.')
        print(Style.RESET_ALL)  # Tells player remaining number of correct guess need to be made.

        select_letter = input('Please guess a letter\n\n')

        letter = str(select_letter.upper()) # To accept both lower and upper case letter entered by player

        # This if conditional section below does the job of warning user not to repeat the rightly guessed 
        # letter again without disturbing the result of the game
        
        if letter in guessed_letter_list:
            print(Back.RED,"Don't repeat your correct guess!\n")
            print(Style.RESET_ALL)
            continue
            
        # This section of conditional statement will keep track of correctly guessed letters and identify them 
        # from wrongly guessed letters
        if letter in pick:
            for i in range(0, len(pick)):

                if letter == pick[i]:
                    game_status_list.insert(i, letter)
                    guessed_letter_list.add(letter)
                    counter_letter_list.append(letter)
                    if game_status_list[i+1] == '  ':
                        game_status_list.pop(i+1)
            print(Fore.LIGHTBLUE_EX,'\n YOUR GAME STATUS\n')
            print(Style.RESET_ALL)
            print(game_status_list,'\n')
            print(Back.MAGENTA, 'Rightly Guessed Letters: ', guessed_letter_list)
            print(Back.CYAN,'Wrongly Guessed Letters: ', wrong_guess_list)
            print(Style.RESET_ALL)

        else:
            counter +=1                          
            wrong_guess_list.append(letter)
            print(Fore.LIGHTBLUE_EX,'\n YOUR GAME STATUS\n')
            print(Fore.RED,'Incorrect selection! Try again!\n')
            print(Style.RESET_ALL)
            print('Hangman Status: ', end = ' ')
            for ind in range(0, counter):
                print('|', end = ' ')
            print('\n\n',game_status_list)
            print(Back.MAGENTA, 'Rightly Guessed Letters: ', guessed_letter_list )
            print(Back.CYAN,'Wrongly Guessed Letters: ', wrong_guess_list)
            print(Style.RESET_ALL)

        # If player enters more the six wrong entry, warning message is displayed and game will restart.
        if counter == 6:
            print('Six incorrect guess! YOU ARE HANGED')
            print(Fore.GREEN,'Right word is:', pick, '\n')
            print(Style.RESET_ALL)
            Game_restart()
            
            
        # If player makes right guess of word before makis six wrong guess this section of code will be initiated
        # and game restarts
        if len(counter_letter_list) == len(game_status_list):
            print(Back.BLUE,'Congratulations!!! YOU GOT IT RIGHT')
            print(Style.RESET_ALL)
            Game_restart()
            counter = 6      # This counter value comes here because we want to escape while loop and begin again.
        
```
