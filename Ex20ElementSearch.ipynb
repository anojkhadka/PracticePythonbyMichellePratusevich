### 2)  Binary Search Algorithm (Iterative Method)

This is a binary search algorithm that derives its concept originally from mathematics where "bisection method"
is used in numerical analysis to find square root of a given number. Here however, we use the concept to search an element in a given list. There are two ways generally we can do some task repeatedly when the previous task fails. One is iterative and another is recursive. This algorithm uses iterative approach to write a binary search function.

```python

def Binary_Search(test_list, test_element):
    
    '''
    
    This function takes a sorted list and any element inside that list as an argument. This function works only
    under following conditions:
    
    1) List must contain uniform data types. Either it must contain only integers or floats or strings. 
    
    2) This function fails for list containing mixed data types. 
    
    3) List in which element is being searched must be sorted out in ascending order. List with uniform data 
       types but not sorted out in ascending order will produce error.
       
    4) If the element is not inside the list this function will return error. 
    
    5) If an element appears more than once in the list, this code will be able to pick only one of them.

     As an output, this function will return the number of loop cycles code took to search the given element and
     index position of that element in the list.
    
    '''
    
    
    lower_bounary = 0
    upper_boundary = len(test_list)
    number_of_iteration = 0

    while True:
        
        index_number = (upper_boundary + lower_bounary)//2
        number_of_iteration += 1                             # Loop Counter

        if test_list[index_number] == test_element:
            print(f'Test element {test_element} lies at index position - {index_number}  in the list. \n\n')
            print(f'It took {number_of_iteration} loop cycles to search the test element')
            break
            
        elif test_list[index_number] < test_element:
            lower_bounary = index_number 
        
        else:
            upper_boundary = index_number
        

    return index_number, number_of_iteration
```

### 3) Binary Search Algorithm (Recursive Method)

This algorithm performs the same function as does the previous one. However the fundamental difference is in the method by which it performs the repetition of task. Instead of going through different loop cycles this function will perform the task of function recursively. Concept behind the algorithm is same. 

```python
def Binary_Recursion(test_list, test_element):
    
    '''
    
    This function takes a sorted list and any element inside that list as an argument. This function works only
    under following conditions:
    
    1) List must contain uniform data types. Either it must contain only integers or floats or strings. 
    
    2) This function fails for list containing mixed data types. 
    
    3) List in which element is being searched must be sorted out in ascending order. List with uniform data 
       types but not sorted out in ascending order will produce error.
       
    4) If the element is not inside the list this function will return error. 
    
    5) If an element appears more than once in the list, this code will be able to pick only one of them.
    
    Since recursion does not run through different cycles this function does not return the number of loop 
    cycles as an output. It only returns the index position of the element in the list.  
    
    '''
    
    global lower_bound, upper_bound

    lower_bound = 0
    upper_bound = len(test_list)

    def index_search():
        
        global lower_bound, upper_bound
        
        index_number = (lower_bound + upper_bound)//2

        if test_list[index_number] == test_element:
            print(f'Test element {test_element} lies at index position - {index_number}  in the list. \n\n')
            return index_number
            
        elif test_list[index_number] < test_element:
            lower_bound = index_number
            return index_search()
    
        else:
            upper_bound = index_number
            return index_search()
                    
        return index_number

    print(index_search())
```
