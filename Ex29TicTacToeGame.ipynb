1. If wrong row column index is entered or player enters the index position that has already been filled, it        will ask the user to enter the right index position again displayed in red.

4. Once all the positions have been filled the game automatically stops.



```python
from colorama  import Style, Fore, Back
```

```python
'''This function is created to ask user to input their row and column index for the position where player 
    want to input their tic tac toe mark. This function will later be used and does not take any input. 
    However, it returns the row and column index entered by the player'''


def row_col_input():
        
    row_col = input('Please enter the position \n')
    place_holder = row_col.strip()                      # Changes string to lists and pick row and column index
    m = int(place_holder[0])                            # Gives index position of the row
    n = int(place_holder[2])                            # Gives index position of the column
    return [m,n]
```

```python
'''This function does the job of printing the square board grid. It takes list as an input argument and returns
   the tic tac toe game display as an output. Later this function will take the sample_list' that contains all
   the tokens in their respective positions and use this 'sample_list' as reference to print the game grid. '''

def game_board_grid(token_list):
    h_unit = '---'                         # fundamental horizontal unit of grid
    v_unit = '|'                           # fundamental vertical unit of grid
  
    # This is the head loop that will cause program to iterate to the number of times of the board size.
    
    for i in range(0, 4):                
        for n in range(0, 3):              # This loop prints horizontal structure
            if n == 0:                     # Inserts a single space in front of each horizontal grid
                print(' ', end = '')
            print(h_unit, end = ' ')
        print()                            # This print statement takes to next line after horizontal printing. 
        if i == 3:                         # This 'if conditional' is used to stop loop after the last vertical
            break                          # grid structure is printed.
           
        for m in range(0,4):               # This branch of loop prints vertical structure in the grid.
            print(v_unit, end = '')
            
            if m <3:                                              # This conditional block will print all the 
                print(' ' + token_list[i][m] + ' ', end = '')    # tokens in their respective positions.

        print()                            # This print statement takes to next line after vertical printing.
```

```python
def Winner_check(test_list):
    '''This function will do the job of determining if any player won the game. To do this, it will check all 
       possible position for which the game can be won. There are 8 possible ways for any player to win the game.
       All those 8 possible conditions are kept in a 'test_conditions' list by indexing the position of all such
       8 ways in which game can be won. This function will take list as an input and return integer number for
       number of wins two player made.'''
    
    test_conditions = [test_list[0], test_list[1],test_list[2], [test_list[0][0], test_list[1][0], test_list[2][0]],
                      [test_list[0][1], test_list[1][1], test_list[2][1]], [test_list[0][2], test_list[1][2], 
                       test_list[2][2]], [test_list[0][0], test_list[1][1], test_list[2][2]],[test_list[0][2], 
                        test_list[1][1], test_list[2][0]] ]

    player_one_wins = ['1', '1', '1']   # All 8 test conditions will be mapped to this list to check if any win.
    player_two_wins = ['2', '2', '2']   # Same as above but for second player.
    
    counter_one = 0                     # Counter to check how many row/column wins player makes in one game
    counter_two = 0                     # Counter for player two
    
    # After each game is over this function will be called and this loop below will test all the possible 
    # conditions to check for the number of wins by both the player.
    
    for m in range(0, len(test_conditions)):

        if test_conditions[m] == player_one_wins:
            counter_one +=1

        elif test_conditions[m] == player_two_wins:
            counter_two +=1
            
    return [counter_one, counter_two]
```
